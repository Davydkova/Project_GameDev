Данные исследуются с 1980 года по 2016 включительно.

Для обработки пропущенных значений в столбце year_of_release исследуем количественную зависимость популярной платформы 

(в то время) и год выпуска игры.    

```python
a=[1980,1985,1990,1995,2000,2005,2010,2016]  #разделим наши периоды на пятилетки
genre_year=[]  #  пустой список 
for i in range(len(a[:-1])):  
    df1=df[(df['year_of_release']>=a[i]) & (df['year_of_release']<a[i+1])].groupby('platform')['year_of_release'].count().sort_values(ascending=False).reset_index() 
    genre_year.append(df1['platform'][0])
    
     # создадим логическое условие в соответствии с  выпуском (годом игры)  и количеством популярных платформ в то время
```

```python
print(genre_year)    # популярные платформы в соответствии с годами списка а 
```

### Замена типа данных 

Проверяем тип данных object в столбцах, он строчный , замена не требуется

Год выпуска игр должен быть целочисленным числом, заменяем с вещественного на целочисленный. 

В оценке пользователей тип object, логичнее сделать вещественный, но в данном столбце содержится аббревиатура 'tbd'-to be done.

В данном столбце это скорее всего значит то, что пользователь не оставил свою оценку на данный момент, но воспользовался продуктом, и может оставить свою оценку позже.

В количественном отношении этих данных более 14 процентов, для удаления без существенных потерь это много. 

Для замены на вещественный тип необходимо заменить 'tbd'  на NaN.
