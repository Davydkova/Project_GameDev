##  Предобработка данных и замена типа данных

```python
df.isnull().mean()   # процентное соотношение пропущенных данных
```

```python
missing_values=df.isnull().sum()[df.isnull().sum()>0].sort_values(ascending=False)
missing_values    #отсортированные столбцы по пропущенным данным 
```

```python
len(df['name'].unique())        # количество уникальных значений в столбце с названием игр
```

Всего 2 пропущенных значения в столбце с играми, скорее всего просто сыграла роль человеческого фактора и их забыли заполнить. 

Уникальное количество названия игр слишком велико чтобы мы смогли корректно заполнить пропущенные значения, но и их всего нечего.

Поэтому безболезненно удаляем пропуски в столбце name

```python
df=df.dropna(subset=['name'])     # удаление пропущенных значений
```

Данные исследуются с 1980 года по 2016 включительно.

```python
df=df[df['year_of_release']!=0]     #перезапишем датафрейм без нулевых значений по году
```

```python
df['year_of_release'].hist()        #проверка распределения на начилие нулевых значений в годах
```

Посмотрев на датасет с условием пропущенных значений в годах и названием игр, мы можем заметить одну такую "хорошую особенность" в некоторых названиях игр уже прописан год их релиза,можно было бы его применить непосредственно к пропущенным значениям в годах релиза игр, но к сожалению не всегда Fifa 2010 это игра выпущенная именно в 2010 году (вспоминается книга Метро 2033)

### Замена типа данных 

Проверяем тип данных object в столбцах, он строчный , замена не требуется

Год выпуска игр должен быть целочисленным числом, заменили ранее, в процессе обработки. 

В оценке пользователей тип object, логичнее сделать вещественный, но в  столбце содержится аббревиатура 'tbd'-to be done, она требует замены, как отдельная группа (например "-1"), если необходима в отдельна в дальнейшем исследовании, иначе замена на пропущенное значение.

В данном столбце это скорее всего значит то, что пользователь не оставил свою оценку на данный момент, но воспользовался продуктом, и может оставить свою оценку позже.

Автоматизация. Создадим функцию для подачи параметров колонок(платформы, рейтинга, жанра) 


```python
def getTops(column):
    return(
  df_sales_unpivot
  .groupby(['country',column])
  .sales
  .sum()
  .reset_index()
  .sort_values(by=['country','sales'],ascending=[True,False])  #сортируем по стране и продажам
  .groupby(['country'])
  .head(5) #выбираеем топ-5 в каждой стране
  .merge(sales_by_country, on ='country') #присоединяем продажи по странам 
  .assign(sales_share=lambda x:(x['sales']/x['total_sales']))  #считаем долю от общего числа продаж в стране 
  .drop('total_sales',axis=1) #удаляем колонку с полными продажами
  .style.format({
      'sales':'{:,.1f}'.format,
  'sales_share':'{:,.1%}'.format
  }) #стиль
)
```

Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

 Пороговое значение alpha=0.05



    Нулевая гипотеза H₀: Средние пользовательские рейтинги жанров Action и Sports одинаковые.
    Альтернативная гипотеза H₁: Средние пользовательские рейтинги жанров Action и Sports различаются.


```python
genre_action_hyp = df_active[df_active['genre']=='Action'] ['user_score']
genre_sports_hyp = df_active[df_active['genre']=='Sports'] ['user_score']
```
